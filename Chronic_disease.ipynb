```python

print(df[numerical_columns].isnull().sum())
# from above plot we can see numerical columns are having outliers
# so the null values in the numerical columns will be replaced by median values wrt to each column

```

```python
median_values = df[numerical_columns].median()
median_values
```

```python
df[numerical_columns] = df[numerical_columns].fillna(median_values)

# so in the age column the null values are replaced with median value of age
```

```python
df[numerical_columns]
```

```python
df[numerical_columns].isna().sum()
```

```python
mode_values = df[categorical_columns].mode().iloc[0]
```

```python
df[categorical_columns].isnull().sum()
```

```python
df[categorical_columns] = df[categorical_columns].fillna(mode_values)
```

```python
# we can see there are few object dtype columns, now we need to convert them into numerical

from sklearn.preprocessing import LabelEncoder

label_Encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = label_Encoder.fit_transform(df[col])
```
